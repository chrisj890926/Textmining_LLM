```python
import fitz  # PyMuPDF
import re

def extract_core_terms_with_stopwords(pdf_file_path):
    with fitz.open(pdf_file_path) as doc:
        full_text = ""
        for page_num in range(len(doc)):
            page_text = doc[page_num].get_text("text")
            full_text += page_text + "\n\n"

    # 根据提供的正则表达式提取“Core Terms”内容，直到遇到指定的停止词
    matches = re.finditer(r'Core Terms\s+([\s\S]+?)(?=\n(?:[A-Z][^\s]*(?:\s[A-Z][^\s]*)*[:\n]|Opinion by:))', full_text, re.MULTILINE)
    
    for case_index, match in enumerate(matches, start=1):
        core_term_content = match.group(1).strip()
        print(f"Case {case_index}, Core Terms:\n{core_term_content}\n")
        print('-----\n')  # 添加分隔符以区分不同的 'Core Terms' 文本

# 示例文件路径，需要替换为您的文件路径
pdf_file_path = r'C:\Users\User\Dropbox\textmining1\PDF\Files (100).PDF'

extract_core_terms_with_stopwords(pdf_file_path)
```

```python
import fitz  # PyMuPDF
import re
import csv
import os
from pathlib import Path
def extract_core_terms_with_stopwords(pdf_file_path, csv_file_path):
    with fitz.open(pdf_file_path) as doc:
        full_text = ""
        for page_num in range(len(doc)):
            page_text = doc[page_num].get_text("text")
            full_text += page_text + "\n\n"

    # 打开CSV文件准备写入
    with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        # 写入标题行
        csv_writer.writerow(["Case Index", "Core Terms"])

        # 根据提供的正则表达式提取“Core Terms”内容，直到遇到指定的停止词
        matches = re.finditer(r'Core Terms\s+([\s\S]+?)(?=\n(?:[A-Z][^\s]*(?:\s[A-Z][^\s]*)*[:\n]|Opinion by:))', full_text, re.MULTILINE)
        
        for case_index, match in enumerate(matches, start=1):
            core_term_content = match.group(1).strip()
            # 将每个匹配的“Core Terms”写入CSV文件
            csv_writer.writerow([f"Case {case_index}", core_term_content])

def process_pdf_in_folder(folder_path, output_csv_folder_path):
    Path(output_csv_folder_path).mkdir(parents=True,exist_ok=True)

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith('.pdf'):
                pdf_file_path = os.path.join(root,file)
                csv_file_name = file.replace('.pdf', '_core_terms.csv').replace('.PDF','_core_terms.csv')
                csv_file_path = os.path.join(output_csv_folder_path, csv_file_name)

                extract_core_terms_with_stopwords(pdf_file_path, csv_file_path)

data1_folder_path = r'C:\Users\User\Dropbox\textmining1\Data 3'
output_csv_folder_path = r'C:\Users\User\Dropbox\textmining1\Data 3' 

process_pdf_in_folder(data1_folder_path, output_csv_folder_path)
```
