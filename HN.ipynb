## 彙整CSV

```python
import csv
import os
from pathlib import Path

def natural_sort_key(s):
    """
    提取用于自然排序的键（包括数字排序）。
    """
    import re
    return [int(text) if text.isdigit() else text.lower() for text in re.split('([0-9]+)', s)]

def merge_csv_files(input_folder, output_csv_path):
    # 确保输出文件夹存在
    Path(output_csv_path).parent.mkdir(parents=True, exist_ok=True)
    
    with open(output_csv_path, 'w', newline='', encoding='utf-8') as output_file:
        output_writer = csv.writer(output_file)
        headers_written = False

        # 获取所有CSV文件路径，并按自然排序
        csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]
        csv_files.sort(key=natural_sort_key)  # 使用自定义排序键
        
        for csv_file in csv_files:
            csv_file_path = Path(input_folder) / csv_file
            pdf_label = csv_file.split('_')[0]  # 文件名格式假设为 'a1_HN.csv'

            with open(csv_file_path, newline='', encoding='utf-8') as input_csv_file:
                csv_reader = csv.reader(input_csv_file)
                headers = next(csv_reader)

                if not headers_written:
                    # 如果还没有写入头部，先写入头部，加上'PDF'列
                    output_writer.writerow(['PDF'] + headers)
                    headers_written = True

                for row in csv_reader:
                    # 为每行数据加上PDF标签
                    output_writer.writerow([pdf_label] + row)

# 指定包含要合并的CSV文件的文件夹路径和输出CSV文件的路径
input_folder = r'C:\Users\User\Dropbox\textmining1\HN'
output_csv_path = r'C:\Users\User\Dropbox\textmining1\merged_HN.csv'

merge_csv_files(input_folder, output_csv_path)
```
