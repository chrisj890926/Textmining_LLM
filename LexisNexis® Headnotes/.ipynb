```python
import fitz  # PyMuPDF
import re
import csv

def extract_cases_and_process(pdf_file_path, csv_file_path):
    cases_text = []

    with fitz.open(pdf_file_path) as doc:
        full_text = ""
        for page_num in range(len(doc)):
            page_text = doc[page_num].get_text("text")
            full_text += page_text + "\n\n"

        case_splits = re.split(r'Core Terms|End of Document', full_text)
        cases_text = [case_splits[i] for i in range(1, len(case_splits), 2)]

    headnotes_counter = 0  # 初始化计数器

    for case_index, case_text in enumerate(cases_text, start=1):
        headnotes = re.findall(r'LexisNexis® Headnotes(.+?)(?=HN\d+|$)', case_text, re.DOTALL)
        
        # 计算并累加Headnotes的数量
        headnotes_counter += len(headnotes)
        
        # 打印或保存提取的信息
        print(f"Case {case_index}, Headnotes:", headnotes)

    # 打印LexisNexis® Headnotes总数量
    print(f"Total LexisNexis® Headnotes Count: {headnotes_counter}")

# 示例文件路径，需要替换为您的文件路径
pdf_file_path = r'C:\Users\User\Dropbox\textmining1\PDF\Files (100).PDF'
csv_file_path = r'C:\Users\User\Dropbox\textmining1\LexisNexis® Headnotes.csv'

extract_cases_and_process(pdf_file_path, csv_file_path)
```
