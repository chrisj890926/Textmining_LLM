```python
import fitz
import re

pdf_file_path = '/Users/chingbochen/Desktop/Files-35.PDF'
txt_file_path = '/Users/chingbochen/Desktop/Files-35-matches1115-1.txt'

pattern = r'(Counsel:.*?|Judges:.*?|Opinion by:.*?)(?=Counsel:|Judges:|Opinion by:|Opinion|$)'

full_text = ""

with fitz.open(pdf_file_path) as doc:
    for page in doc:
        full_text += page.get_text()

matches = re.findall(pattern, full_text, re.DOTALL)

with open(txt_file_path, 'w') as txt_file:
    for match in matches:
        txt_file.write(match.strip() + "\n\n")

print(f"完成：{txt_file_path}")
```
