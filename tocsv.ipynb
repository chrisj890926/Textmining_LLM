```python
import json
import csv

# 讀取 TXT 檔案（假設它是 JSON 格式）
with open("Matched_ALL_DATA.txt", "r", encoding="utf-8") as file:
    data = json.load(file)  # 解析 JSON

# 定義輸出 CSV 檔案
output_file = "output.csv"

# 開啟 CSV 檔案進行寫入
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    # 寫入 CSV 標題
    writer.writerow(["Case Index", "Context", "Name", "Definition"])

    # 遍歷 JSON 數據
    for entry in data:
        case_index = entry.get("Case Index", "")
        context = entry.get("Context", "")

        # 確保 Knowledge Graph Result 存在且為列表
        knowledge_results = entry.get("Knowledge Graph Result", [])
        for knowledge in knowledge_results:
            name = knowledge.get("Name", "")
            definition = knowledge.get("Definition", "")

            # 寫入 CSV，每個 Name-Definition 為一行
            writer.writerow([case_index, context, name, definition])

print(f"轉換完成！CSV 檔案已儲存為 {output_file}")
```

```python
import csv
import re

# 讀取 TXT 檔案
with open("/Users/tangjiahong/Dropbox/textmining1/Matched_txt/Matched_ALL_DATA.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# 初始化變數
output_data = []
current_case_index = None
current_context = None
current_name = None
current_definition = None

# 正則表達式模式
case_pattern = re.compile(r'"Case Index":\s*(.+)')  # 提取 Case Index
context_pattern = re.compile(r'"Context":\s*"(.+)"')  # 提取 Context
name_pattern = re.compile(r'"Name":\s*"(.+)"')  # 提取 Name
definition_pattern = re.compile(r'"Definition":\s*"(.+)"')  # 提取 Definition

# 遍歷文件內容
for line in lines:
    line = line.strip()  # 去除前後空白
    
    # 匹配 Case Index
    match = case_pattern.search(line)
    if match:
        current_case_index = match.group(1)
        continue

    # 匹配 Context
    match = context_pattern.search(line)
    if match:
        current_context = match.group(1)
        continue

    # 匹配 Name
    match = name_pattern.search(line)
    if match:
        current_name = match.group(1)
        continue

    # 匹配 Definition
    match = definition_pattern.search(line)
    if match:
        current_definition = match.group(1)

        # 儲存數據
        output_data.append([current_case_index, current_context, current_name, current_definition])

# 定義輸出 CSV 檔案
output_file = "output.csv"

# 寫入 CSV
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Case Index", "Context", "Name", "Definition"])  # 標題
    writer.writerows(output_data)

print(f"轉換完成！CSV 檔案已儲存為 {output_file}")
```
