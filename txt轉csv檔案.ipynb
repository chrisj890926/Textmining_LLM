```python
import re
import csv


def extract_info(text):

    counsel_pattern = r"Counsel:(.*?)(?=\n\w+:)"
    judges_pattern = r"Judges:(.*?)(?=Opinion by:)"
    opinion_by_pattern = r"Opinion by:(.*?)(?=\n\w+:|\Z)"

    counsels = re.findall(counsel_pattern, text, re.DOTALL)
    judges = re.findall(judges_pattern, text, re.DOTALL)
    opinions_by = re.findall(opinion_by_pattern, text, re.DOTALL)

    return counsels, judges, opinions_by


file_path = '/Users/chingbochen/Desktop/Files-35-matches1115.txt'
```

```python
with open(file_path, 'r') as file:
    content = file.read()


counsels, judges, opinions_by = extract_info(content)


csv_file_path = '/Users/chingbochen/Desktop/Files-35-matches.csv'
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Writing the header
    writer.writerow(['Counsel', 'Judges', 'Opinion By'])
    # Writing the data
    for i in range(len(counsels)):
        writer.writerow([counsels[i].strip(), judges[i].strip(), opinions_by[i].strip() if i < len(opinions_by) else ""])
print('完成！')
```
